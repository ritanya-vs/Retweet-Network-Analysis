```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## **JUSTIN BIEBER**

```python
justin=pd.read_csv('justinbieber_with_timestamp.csv')

jedges = justin[['Node1', 'Node2']].values.tolist()


jgraph = nx.from_edgelist(jedges)


nx.draw_networkx(jgraph,nx.spring_layout(jgraph),with_labels=False)
plt.show()
```

The graph helps to understand the overall structure and connections between users involved in retweeting hashtags related to Justin Bieber.

```python
jn=jgraph.number_of_nodes()
je=jgraph.number_of_edges()
jd=jgraph.degree()
print('Number of nodes : ',jn)
print('Number of edges : ',je)
print('Degree of nodes : ',jd)
```

In this network, node 722 has the highest degree, with 3,875 connections, indicating that it plays a central role in the network .
This high degree suggests significant connectivity,serving as a major hub in information spread within the dataset.

```python
# Function for temporal analysis of retweet networks grouped by week
def temporal_analysis_by_week(df, title):
    # Check if the DataFrame is empty
    if df.empty:
        print("The DataFrame is empty. Cannot perform temporal analysis.")
        return

    # Convert timestamp to datetime format
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
    df['year_week'] = df['Timestamp'].dt.to_period('W')  # Group by year and week

    # Group data by each week
    weekly_groups = df.groupby('year_week')

    # Determine grid size based on the number of weeks
    num_weeks = len(weekly_groups)
    if num_weeks == 0:
        print("No data available for weekly analysis.")
        return

    num_cols = 3
    num_rows = (num_weeks // num_cols) + (num_weeks % num_cols > 0)

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
    fig.suptitle(title, fontsize=16)

    # Flatten axes array for easy iteration, even if it's a single row
    axes = axes.flatten() if num_weeks > 1 else [axes]

    # Iterate over each week and create a subplot
    for i, (week, group) in enumerate(weekly_groups):
        ax = axes[i]

        # Create a graph for each week's data
        G_week = nx.from_pandas_edgelist(group, 'Node1', 'Node2', create_using=nx.Graph())

        # Get position for all nodes in the weekly network
        pos = nx.spring_layout(G_week)

        # Draw network on the subplot
        nx.draw_networkx_nodes(G_week, pos, ax=ax, node_size=50, node_color='blue', alpha=0.7)
        nx.draw_networkx_edges(G_week, pos, ax=ax, edge_color='gray', alpha=0.5)
        ax.set_title(f'Week: {week}', fontsize=12)
        ax.axis('off')

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)  # Adjust to make space for the main title
    plt.show()

```

```python
# Apply temporal analysis by week to both datasets
print("Temporal Analysis for Justin Bieber Network (by week):")
temporal_analysis_by_week(justin, "Justin Bieber Network (Weekly Analysis)")
```

For Justin Bieber dataset , in the initial week a larger community is invloved in retweeting , in the further weeks we can see there are more retweets in smaller communities.

```python

print("Temporal Analysis for One Direction Network (by week):")
temporal_analysis_by_week(one, "One Direction Network (Weekly Analysis)")
```

```python
def analyze_information_spread(df):
    # Convert timestamp to datetime format
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
    df['year_week'] = df['Timestamp'].dt.to_period('W')

    # Group data by each week
    weekly_groups = df.groupby('year_week')

    # Store metrics for analysis
    spread_metrics = []

    # Iterate over each week
    for week, group in weekly_groups:
        # Create the graph for the week
        G_week = nx.from_pandas_edgelist(group, 'Node1', 'Node2', create_using=nx.DiGraph())

        # Calculate network metrics
        num_nodes = G_week.number_of_nodes()
        num_edges = G_week.number_of_edges()
        largest_component_size = len(max(nx.weakly_connected_components(G_week), key=len))
        avg_clustering_coefficient = nx.average_clustering(G_week.to_undirected())

        # Measure centrality metrics to identify key influencers
        centrality = nx.degree_centrality(G_week)
        max_centrality = max(centrality.values()) if centrality else 0

        # Store metrics for the week
        spread_metrics.append({
            'Week': str(week),
            'Num_Nodes': num_nodes,
            'Num_Edges': num_edges,
            'Largest_Component_Size': largest_component_size,
            'Avg_Clustering_Coefficient': avg_clustering_coefficient,
            'Max_Centrality': max_centrality
        })

    # Convert metrics to DataFrame for further analysis
    metrics_df = pd.DataFrame(spread_metrics)
    print(metrics_df)

    # Plot some key metrics over time
    plt.figure(figsize=(12, 6))
    plt.plot(metrics_df['Week'], metrics_df['Num_Nodes'], label='Number of Nodes', marker='o')
    plt.plot(metrics_df['Week'], metrics_df['Num_Edges'], label='Number of Edges', marker='s')
    plt.plot(metrics_df['Week'], metrics_df['Largest_Component_Size'], label='Largest Component Size', marker='^')
    plt.xticks(rotation=45)
    plt.xlabel('Week')
    plt.ylabel('Metric Value')
    plt.title('Weekly Information Spread Metrics')
    plt.legend()
    plt.tight_layout()
    plt.show()

    return metrics_df
```

```python
# Analyze information spread in the network
metrics_df = analyze_information_spread(justin)
```

The network is large, with 5,151 nodes and 5,230 edges, and the largest component includes nearly all nodes (5,122), suggesting a well-connected network that can facilitate broad information spread. The average clustering coefficient is low (0.001452), indicating minimal local clustering, or "grouping," among nodes. Despite this, the high max centrality (0.715) suggests that a few highly connected nodes act as major hubs, driving network-wide connectivity and effective information dissemination.

In Weeks 1 and 2, the network is large and well-connected, with highly central nodes that enable effective information spread despite low local clustering. By Week 3, the network becomes highly fragmented, with only a few connected nodes, no central hubs, and no clustering, greatly reducing its ability to facilitate widespread information dissemination.

```python
def calculate_diffusion_speed(df, target_users):
    # Convert timestamp to datetime
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
    df = df.sort_values('Timestamp')  # Sort by time to track diffusion

    # Track unique users over time
    unique_users = set()
    time_to_reach = []

    for i, row in df.iterrows():
        unique_users.add(row['Node2'])  # Assuming 'Node2' is the user who retweeted
        if len(unique_users) >= target_users:
            # Record the time it took to reach the target number of users
            time_to_reach.append(row['Timestamp'])
            break

    if time_to_reach:
        # Calculate the time difference
        start_time = df['Timestamp'].min()
        time_taken = (time_to_reach[0] - start_time).total_seconds() / 3600  # Convert to hours
        print(f"Time to reach {target_users} users: {time_taken:.2f} hours")
    else:
        print(f"Did not reach {target_users} users within the data.")

def calculate_information_flow_rate(df, time_interval='1H'):
    # Convert timestamp to datetime and sort
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
    df = df.sort_values('Timestamp')

    # Group by specified time interval
    retweets_per_interval = df.groupby(pd.Grouper(key='Timestamp', freq=time_interval)).size()

    # Plot the information flow rate
    plt.figure(figsize=(12, 6))
    retweets_per_interval.plot(kind='line', marker='o', label='Retweets per Interval')
    plt.xlabel('Time')
    plt.ylabel('Number of Retweets')
    plt.title('Information Flow Rate Over Time')
    plt.legend()
    plt.tight_layout()
    plt.show()

    return retweets_per_interval
```

```python
# Calculate diffusion speed for reaching 100 users
calculate_diffusion_speed(justin, target_users=100)

# Calculate and plot information flow rate (retweets per hour)
flow_rate = calculate_information_flow_rate(justin, time_interval='1h')
```

Given that it takes about 20.77 hours to reach 100 users, the information flow rate can be calculated as the number of users gained per hour. This indicates how quickly information is spreading through the network.
